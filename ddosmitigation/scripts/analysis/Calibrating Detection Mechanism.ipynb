```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
sw11 = np.loadtxt('sw11.txt')
sw21 = np.loadtxt('sw21.txt')
sw41 = np.loadtxt('sw41.txt')
sw31 = np.loadtxt('sw31.txt')
sw51 = np.loadtxt('sw51.txt')
```

```python
print ('####  SWITCH 11  ####')
       
src11 = sw11[:,1]
dst11 = sw11[:,2]

long11 = len(src11)
index11 = range(1,long11+1)

plt.figure(figsize=(8, 5))
plt.scatter(index11,src11 / 16, color='black', lw=0.5, label="Source Entropy - SW11")
plt.xlabel('Observation Window')
plt.ylabel('Source Entropy Value')
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(index11,dst11 / 16, color='black', lw=0.5, label="Destination Entropy - SW11")
plt.xlabel('Observation Window')
plt.ylabel('Destination Entropy Value')
plt.show()

msrc11 = np.mean (src11)
stdsrc11 = np.std (src11)

mdst11 = np.mean (dst11)
stddst11 = np.std (dst11)

print ('SW11 - Entropy Source Mean',msrc11,'Entropy Source Standard Deviation',stdsrc11)
print ('SW11 - Entropy Destination Mean',mdst11,'Entropy Destination Standard Deviation',stddst11)
```
